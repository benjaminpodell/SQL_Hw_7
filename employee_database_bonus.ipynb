```python
# Importing Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sqlalchemy import create_engine
```

```python
# Engine variable to pass in postgres connection with my password in order to connect to database
engine_var = create_engine('postgres://postgres:postgres@localhost:5432/Employee_Database')

# Variable to connect to engine
bridge = engine_var.connect()
```

```python
# Variable that queries and reads all data from the 'salaries' table and pass conn variable for the connection
sql_to_pd_salaries = pd.read_sql("SELECT * FROM salaries", bridge)

# Variable that queries and reads all data from the 'titles' table and pass conn variable for the connection
sql_to_pd_titles = pd.read_sql("SELECT * FROM titles", bridge)

# Variable that combined both tables then displays the first 10 rows
salaries_and_titles = pd.merge(sql_to_pd_salaries, sql_to_pd_titles, on="emp_no", how="inner")
salaries_and_titles.head(10)
```

```python
# Grouping the 'salaries_and_titles_grouped' by title and taking the average of each
salaries_and_titles_grouped = salaries_and_titles.groupby("title").mean()

# Variable to store the grouped dataframe reset indexes then displaying the dataframe
indexed_df = salaries_and_titles_grouped.reset_index()
indexed_df
```

```python
# Variables that store x_axis, y_axis, and ticks from 'title' and 'salary'
x_axis = indexed_df["title"]
y_axis = indexed_df["salary"]
ticks = np.arange(len(x_axis))

# Creates the bar chart and setting color
plt.bar(x_axis, y_axis, alpha=0.8, color="c")

# Creates ticks for the x axis and displays titles vertically 
plt.xticks(ticks, x_axis, rotation=90)

# Creates bar graph title, x label, and y label
plt.title("Average Salary by Title")
plt.xlabel("Title")
plt.ylabel("Salary ($)")

# Saves an image of the bar graph
plt.savefig("Average Salary by Title.png")

# Shows the graph
plt.show()
```
